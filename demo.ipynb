```python
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

import openai
from data_generation import generate_examples
```

```python
def plot_images(images):
    fig, axes = plt.subplots(nrows=1, ncols=len(images))
    for imp, ax in zip(images, axes.ravel()):
        img = mpimg.imread(imp)
        ax.imshow(img)
        ax.axis('off')
    fig.tight_layout()
```

```python
openai.api_key = os.getenv("OPENAI_API_KEY")
```

Add some subjects you'd like to simulate training data for. Some suggestions have been added, but try out others!

```python
subjects = ['computer mouse', 'brown dog', 'baseball'] 
```

```python
images, masks = [], []
for subject in subjects:
    print(f"Generating an example of a {subject}.")
    img, mask = generate_examples(subject=subject, number_of_examples=1)
    images.append(img[0])
    masks.append(mask[0])
```

Now we can visualize the new images we've generated:

```python
plot_images(images)
```

Along with the segmentation masks, that can be used as ground truth:

```python
plot_images(masks)
```
